```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(1,20)
y=2*x+10
plt.title("Matplotlib")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,"--")
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6,6]
plt.plot(x)
```

```python
x=np.arange(10)
plt.plot(x,[y**2 for y in x])
plt.show()
```

```python
x=np.arange(0,1,0.1)
plt.plot(x,[y**2 for y in x])
plt.show()
```

```python
a=np.arange(10)
plt.plot(x,[x**2 for x in a])
plt.plot(x,[x**3 for x in a])
plt.plot(x,[x**1 for x in a])
plt.show()
```
